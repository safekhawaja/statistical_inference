Standard Deviation = root(E(X^2) - E(X)^2)

### b)

Theoretical values: mean = 1.15, SD/sqroot(5) = 0.0666. As we can see, calculated mean and sd (which change everytime the program is run) is really close to the theoretical values above (e.g. 1.1624 and 0.6833).

```
import numpy as np
import matplotlib.pyplot as plt

n = 5 
trial = 0
sim = 500
means = []

while trial < sim:
  sample = np.random.choice(6, size = n, replace = True, p = [0.5, 0.2, 0.1, 0.1, 0.05, 0.05])
  means.append(np.mean(sample))
  trial = trial + 1

print(np.mean(means))
print(np.std(means))
fig, ax = plt.subplots(figsize =(10, 7)) 
ax.hist(means, bins = 15) 
plt.show()
```

### c)

We can see that there is a large right skew from the histogram (particularly above around 1.75 depending on the run). The sample size, however, is not enough for the CLT to have an effect and, when we run the experiment for n = 10 below, we can see that we receive a less skewed distribution that looks more akin to a normal distribution. 

```
n = 10
trial = 0
sim = 500
means = []

while trial < sim:
  sample = np.random.choice(6, size = n, replace = True, p = [0.5, 0.2, 0.1, 0.1, 0.05, 0.05])
  means.append(np.mean(sample))
  trial = trial + 1

print(np.mean(means))
print(np.std(means))
fig, ax = plt.subplots(figsize =(10, 7)) 
ax.hist(means, bins = 15) 
plt.show()
```
